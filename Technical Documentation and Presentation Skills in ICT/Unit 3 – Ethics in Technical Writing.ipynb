# Ethics in Technical Writing

Ethics in technical writing refer to the principles that guide writers in presenting information truthfully, accurately, and responsibly.

## Core Principles of Ethical Writing

- **Truthfulness**: Avoid exaggeration or omission that could lead to misunderstandings.
- **Proper Credit**: Acknowledge all sources, contributors, and original creators.
- **Confidentiality**: Handle sensitive information appropriately and securely.
- **Objectivity**: Minimize bias and present fact-based, balanced content.

## Ethical Dilemmas in Technical Writing

| **Ethical Issue**    | **Example Scenario**                                      | **Consequences**                         |
|----------------------|------------------------------------------------------------|------------------------------------------|
| Misrepresentation     | Overstating software capabilities                         | Loss of user trust, potential lawsuits    |
| Data Falsification    | Altering research results to match expectations           | Scientific misconduct                    |
| Concealment           | Omitting known security risks in documentation            | Ethical and legal violations             |


# Plagiarism in Technical Writing

Plagiarism occurs when someone presents another person's work or ideas as their own without proper attribution.

## Types of Plagiarism

- **Direct Copying**: Using text verbatim without citation.
- **Paraphrasing Without Credit**: Rewriting someone else’s ideas without acknowledgment.
- **Self-Plagiarism**: Reusing your previously published work without disclosure.
- **Improper Citation**: Failing to cite sources accurately.

## Preventing Plagiarism

1. **Use Citation Styles**: Follow IEEE, APA, MLA, or other standard formats.
2. **Check for Plagiarism**: Use tools like Turnitin or Grammarly.
3. **Paraphrase Properly**: Preserve the original meaning and always credit the source.

### Citation Example

**Original Source**:  
> "The usability of software largely depends on how well the documentation is written and structured." – Dr. Jane Doe, *Technical Writing Journal*, 2020.

- **Improper**:  
  The usability of software depends on well-structured documentation.

- **Proper**:  
  As Dr. Jane Doe (2020) states, "The usability of software largely depends on how well the documentation is written and structured."


# Image Ethics in Technical Writing

Visuals such as graphs, screenshots, and diagrams must accurately represent information.

## Unethical Image Practices

- **Editing Data**: Changing chart scales or removing data to mislead.
- **Altering Screenshots**: Modifying UIs to hide issues or exaggerate functionality.
- **Manipulating Photos**: Retouching images to distort the truth.

## Guidelines for Ethical Image Use

| **Ethical Principle** | **Best Practice**                                           |
|-----------------------|-------------------------------------------------------------|
| Accuracy              | Data visuals must reflect true findings.                   |
| Transparency          | Disclose when an image has been modified.                  |
| Attribution           | Cite sources for all visuals.                              |
| No Deception          | Avoid edits that alter the intent or mislead viewers.      |

### Example: Ethical vs. Unethical Image Use

- **Ethical**:  
  A company includes a disclaimer that a product image has been digitally enhanced for demonstration purposes.

- **Unethical**:  
  A marketing team removes error messages from a software screenshot to falsely present the product as more reliable.
